```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urlparse, urlunparse, urljoin
from urllib.robotparser import RobotFileParser
from concurrent.futures import ThreadPoolExecutor,as_completed
from threading import Lock
import requests

import heapq
import time
```

```python
def canonicalize_url(url, base_url=None):
    # Parse the URL normally
    parsed_url = urlparse(url)

    # If it's a relative URL, make it absolute using the provided base URL
    if not parsed_url.netloc and base_url:
        url = urljoin(base_url, url)
    #print(url)
    # Parse the URL again after making it absolute if needed
    #parsed_url = urlparse(url)

    # Convert scheme and host to lower case
    scheme = parsed_url.scheme.lower()
    netloc = parsed_url.netloc.lower()

    # Remove port if it's present
    netloc_parts = netloc.split(':')
    netloc = netloc_parts[0]

    # Remove fragment
    fragment = ''

    # Remove duplicate slashes
    path = parsed_url.path.replace('//', '/')

    # Construct the canonicalized URL
    canonicalized_url = urlunparse((scheme, netloc, path, parsed_url.params, parsed_url.query, fragment))

    # Add the base URL to the canonicalized URL if it's not empty
    if canonicalized_url:
        url_with_base = urljoin(base_url, canonicalized_url)
        return url_with_base

    # If canonicalized URL is empty, return None
    return None

# Example usage:
canonicalized_absolute_url = canonicalize_url("wiki/Kyshtym_disaster", base_url="https://en.wikipedia.org")
print(canonicalized_absolute_url)


# Example usage:
urls = [
    'HTTP://www.Example.com/SomeFile.html',
    'http://www.example.com:80',
    'http://www.example.com:443',
    'http://www.example.com:999',
    'http://www.example.com/a/b.html',
    'http://www.example.com/a.html#helloWorld',
    'http://www.example.com//a.html',
    '../../about.html'
]

base_url = 'http://www.example.com/products/index.html'

for url in urls:
    canonicalized_url = canonicalize_url(url, base_url)
    print(canonicalized_url)
```

```python
def robotParsing(url):
    # Create a RobotFileParser object
    robot_parser = RobotFileParser()
    # Set the URL of the robots.txt file
    robot_parser.set_url(url)
    # Read the robots.txt file
    robot_parser.read()
    # Check if the URL is allowed to be fetched
    return robot_parser.can_fetch("*", url)

# Example
canonicalized_url = "https://en.wikipedia.org/wiki/Kyshtym_disaster"
canonicalized_url1 = "https://example.com/private/page.html"
canonicalized_url2 = "https://www.amazon.com/"
print(robotParsing(canonicalized_url))
print(robotParsing(canonicalized_url1))
print(robotParsing(canonicalized_url2))
```

```python
def fetch_html_content(url):
    
    response = requests.get(url)
    return response

def pageValidity(response):
    headers = response.headers
    lang = headers["content-language"] if "content-language" in headers else "en"
    contentType = headers["content-type"].split(";") if "content-type" in headers else "text/html"
    return lang == "en" and contentType[0] and contentType[0] == "text/html"

def parse_content(response):
    html_content = response.text
    soup = BeautifulSoup(html_content, "html.parser")
    return soup

def extract_title(soup):
    title_tag = soup.find("title")
    document_title = title_tag.get_text() if title_tag else ""
    return document_title

'''def extract_links(soup):
    links = []
    for paragraph_tag in soup.find_all("p"):
        for link_tag in paragraph_tag.find_all("a", href=True):
            #print(link_tag["href"])
            link = canonicalize_url(link_tag["href"],"https://en.wikipedia.org")
            if link:
                links.append(link)
    return links'''

def extract_paragraph_contents(soup):
    paragraph_tags = soup.find_all("p")
    p_contents_list = [paragraph_tag.get_text() for paragraph_tag in paragraph_tags]
    all_p_contents = "\n".join(p_contents_list)
    return all_p_contents

def wrap_content(url, document_title, all_p_contents):
    modified_content = f"<DOCNO>{url}</DOCNO>\n<HEAD>{document_title}</HEAD>\n<TEXT>\n{all_p_contents}\n</TEXT>"
    return modified_content

def extract_contents(url):
    html_content = fetch_html_content(url)
    soup = parse_content(html_content)
    document_title = extract_title(soup)
    #links = extract_links(soup)
    all_p_contents = extract_paragraph_contents(soup)
    modified_content = wrap_content(url, document_title, all_p_contents)
    return modified_content


# Example usage:
#url = "https://en.wikipedia.org/wiki/Kyshtym_disaster"
#content, links = extract_contents(url)
#print("Modified Content:")
#print(content)
#for link in links:
    print(link)
```

```python
# Storing the extracted content into a text file
def writeFile(filename, modified_content):
    with open(output_filename, "w", encoding="utf-8") as file:
        file.write(modified_content) 
```

```python
def traverseLinks(baseUrl, soup, waveNumber,inlinks,outlinks,pq,inlinksCounter):
    # Find all elements of anchor tag
    links = soup.find_all('a', href=True)

    if baseUrl not in outlinks:
        outlinks[baseUrl] = set()
    
    for link in links:
        hrefLink = link.get("href")
        cl = canonicalize_url(hrefLink, baseUrl)
        
        if cl not in inlinks:
            inlinks[cl] = set()
            inlinksCounter[cl] = 0
        
        ## from baseUrl, link is going out
        outlinks[baseUrl].add(cl)
        
        ## from link, baseUrl is coming in
        inlinks[cl].add(baseUrl)
        inlinksCounter[cl] += 1
        
    for cl in inlinksCounter: 
        pq.push(cl, inlinksCounter[cl], waveNumber + 1)
```

```python
class PriorityQueue:
    def __init__(self):
        self._queue = []
    
    # min heap is maintained by default, I beleive this is not expected
    def push(self, link, waveNumber, inLinkCount):
        """
        # Suggestion : waveNumber, -inLinkCount, link
        (4,-3,url1) (4,-2,url2), (2,-10,url3)
        required order : (2,10,url3), (4,-3,url2), (4,-2,url1)
        """
        heapq.heappush(self._queue, (waveNumber, -inLinkCount, link))
    
    def pop(self):
        waveNumber,inLinkCount, link = heapq.heappop(self._queue)
        return waveNumber, link
```

```python
visited_urls= {}
lock=Lock()
def crawl_seed_URL(baseUrl,file_id):
    global visited_urls
    inlinks = {}
    inlinksCounter = {}
    outlinks = {}

    pq = PriorityQueue()
    pq.push(baseUrl, 0, 0)
    now = time.time()

    linksCount =1 
    data=""

    # Set to 40,000 / total_seed_urls you have i.e. 40,000 / 4 = 10,000
    while linksCount <= 10:
        time.sleep(2)
        waveNumber, currentUrl = pq.pop()
        
        
        with lock:
            if currentUrl not in visited_urls:
                # To verify the links and create more seeds to improve speed
                print(f'Link count :{linksCount}, for thread : {file_id}, scanning url: {currentUrl}')
                visited_urls[currentUrl] = 1
            else:
                continue
            
        response = fetch_html_content(currentUrl)
        
        ## if current page is non-html or not in english
        if not pageValidity(response):
            continue
        soup = parse_content(response)
        
        title = extract_title(soup)
        text = extract_paragraph_contents(soup)
        
        traverseLinks(currentUrl, soup, waveNumber,inlinks,outlinks,pq,inlinksCounter)
        data+=wrap_content(currentUrl,title,text)

        # Decide the number to which we want to write to file, default value can be decided as 1000
        # file_ids for each thread will then have a gap of 10
        if linksCount%10==0:
            print(f"Writing files to webpage_data_{file_id}.txt")
            writeFile(f"./webpage_data_{file_id}.txt", data)
            file_id+=1
            data=""
        
        linksCount += 1
        
    print(f'Time in {time.time() - now} seconds')
```

```python
"""
Each seed will initiate a different thread, and file_id is the id of the file where the results will be stored, set the file_id unique enough so that multiple threads don't write onto the same file

"""
seed_map={
    "http://www.world-nuclear.org/info/Safety-and-Security/Safety-of-Plants/Fukushima-Accident/":1,
    "http://en.wikipedia.org/wiki/Fukushima_Daiichi_nuclear_disaster":2,
    "https://thebulletin.org/2019/11/an-update-from-fukushima-and-the-challenges-that-remain-there/":3,
    "https://www.iaea.org/newscenter/focus/fukushima/status-update":4,
}
```

```python
with ThreadPoolExecutor(max_workers=4) as thread_executor:
    for baseUrl,file_id in seed_map.items():
        thread_executor.submit(crawl_seed_URL,baseUrl,file_id)
```
